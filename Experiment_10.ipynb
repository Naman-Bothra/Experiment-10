```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

```python
# Create DataFrame
data = {
    "Name":["a","b","c"],
    "Marks":[85,90,70]
}
df = pd.DataFrame(data)
print(df)
```

```python
# structure of data frame i.e. (column name,shape,size,dimension)
print(df.shape)
print(df.ndim)
print(df.size)
print(df.columns)
```

```python
# check data type of each column
print(df.dtypes)
```

```python
# access individual columns
df["Name"]
```

```python
# Access name of first row
print(df.loc[0,"Name"])
print(df.loc[0,"Marks"])
print(df.iloc[2,1])
```

```python
# add a new column
df["Grade"] = ["First Class","Distinction","Second class"]
print(df)
```

```python
# Update data in datarame
# change marks of student to 88
df.iloc[0,1]=88
print(df)
```

```python
# delete a column
df.drop("Grade",axis=1,inplace=True)
print(df)
```

```python
# perform basic statistical analysis
mean1 = df["Marks"].mean()
print(mean1)
```

```python
min1 = df["Marks"].min()
print(min1)
```

```python
max1 = df["Marks"].max()
print(max1)
```

```python
# apply condition to print values from dataframe

df[df["Marks"]>80]
```
